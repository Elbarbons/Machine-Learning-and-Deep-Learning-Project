```python
from pathlib import Path
import numpy as np
from PIL import Image
from torch.utils.data import Dataset

#PLEASE NOTICE
#the following values are hardcoded and they must be edited before running
#the program in any new environment
TEST_DATASET_ROOT = '/content/Datasets/tokyo_xs/test/'
CSV_ROOT =  '/content/drive/MyDrive/Colab_Notebooks/Dataframes/'

test_path_obj = Path(TEST_DATASET_ROOT)

if not test_path_obj.exists():
    raise Exception(f'Please make sure the path {TEST_DATASET_ROOT} to Tokyo test dataset is correct')


class TokyoTestDataset(Dataset):
    def __init__(self, input_transform = None):


        self.input_transform = input_transform

        # reference images names
        self.dbImages = np.load(CSV_ROOT+'tokyo_xs_test_database.npy')

        # query images names
        self.qImages = np.load(CSV_ROOT+'tokyo_xs_test_queries.npy')

        # ground truth
        self.ground_truth = np.load(CSV_ROOT+'tokyo_xs_test_gt.npy', allow_pickle=True)

        # reference images then query images
        self.images = np.concatenate((self.dbImages, self.qImages))

        self.num_references = len(self.dbImages)
        self.num_queries = len(self.qImages)

    def __getitem__(self, index):
        if index < self.num_references:
            img = Image.open(TEST_DATASET_ROOT + "database/" + self.images[index])
        else:
            img = Image.open(TEST_DATASET_ROOT + "queries/" + self.images[index])

        if self.input_transform:
            img = self.input_transform(img)

        return img, index

    def __len__(self):
        return len(self.images)
```
